and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
from openai import OpenAI
from IPython.display import Markdown, display, update_display
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
MODEL = MODEL_LLAMA
```

```python
# set up environment
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

```python
# here is the question; type over this to ask something new

system_promt = """
You are an eperienced, principle engineer level coder, with expertise on python,
you will answer any question in the previously said domain with great ease-of-understanding and expertise.
Keep it concise and on-point aswell.
you will be provided with any python-related question which you will answer, in plain text, well formatted.
"""


question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get llama3.2 to answer, with streaming
def get_answer_stream(user_prompt):
    stream = ollama.chat.completions.create(
        model=MODEL,
        messages=[
            {'role': 'system', 'content': system_promt},
            {'role': 'user', 'content': user_prompt}
        ],
        stream=True,
    )
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)

get_answer_stream(question)
```
